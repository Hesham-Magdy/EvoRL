- Configs setup

```python
params = """
- iterations: 1000
  algorithm:
    name: ARS
    population: 1024
  env:
    name: Ant
    episode_length: 1000
  network:
    name: MLP
    layers: 64~64~64~64
    hidden_activation: tanh
    output_activation: tanh
  train:
    test_interval: 50
    n_tests: 128
  log:
    log_interval: 1
    visualize_final_env: true
    visualize_test_envs: true
  out_dir: ./results
"""
with open("input.yaml", "w") as f:
    f.write(params)
```
